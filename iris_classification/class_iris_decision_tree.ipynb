```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
columns = ['SepalLengthCm', 'SepalWidthCm','PealLengthCm', 'PetalWidthCm', 'Species']
```

```python
iris = pd.read_csv(url, names = columns)
iris
```

```python
iris.isnull().any()
```

```python
iris.describe()
```

```python
sns.pairplot(iris, hue='Species')
```

<p>Since there is only one dataset available (no separated training and test dataset) we need to divide the dataset into training and test dataset by ourself. To do this, we can use the train_test_split method from the scikit-learn. Don’t forget to split the input and output column to different arrays.</p>

```python
inputs = iris[columns[0:4]].values
inputs
```

```python
classes = iris['Species'].values
classes
```

```python
(train_input, test_inputs, train_classes, test_classes) = train_test_split(inputs, classes, train_size=0.7, random_state = 1)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(train_input, train_classes)
dtc.score(test_inputs, test_classes)
```

```python
dtc.predict(np.array([5, 3, 2, 2]).reshape(1,-1))
```
